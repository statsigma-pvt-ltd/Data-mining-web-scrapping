```python
#some times some web sites will not give access ( 403[forbidden])
# so use below code it will provide information of your fake browser visit.
from urllib.request import urlopen, Request
headers = {'User-Agent': 'Mozilla/5.0'}
#if site doesnt load then change User-agent : "Chrome/41.0.2228.0"
reg_url = "https://www.sciencedirect.com/"
req = Request(url=reg_url, headers=headers) 
html = urlopen(req).read() 
print(html) 
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html.parser")
```

```python
print(soup.prettify())
```

```python
#for getting all paragraph from the site
y=soup.find_all("p")
for x in range(len(y)):
    print(soup.find_all('p')[x].get_text())
```

```python

z=soup.find_all("a")
for x in range(len(z)):
    print(soup.find_all('a')[x].get_text())
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python

soup.title.get_text()
```

```python

Con = "".join([p.text for p in soup.find_all("span",lines=4)])
print(Con)
```

```python
#article from health-sciences
abc=soup.find("div",id="health-sciences")
#abcd=abc.find_all("span",lines=4).....it will not get texts
#below code will combine all string and give text
Con = "".join([p.text for p in abc.find_all("span",lines=4)])
print(Con)
#print(abcd)
```

```python
#article from Social Sciences and Humanities
abc=soup.find("div",id="social-sciences-and-humanities")
#abcd=abc.find_all("span",class_="anchor-text").....it will not get text
#below code will combine all string and give text
Con = "".join([p.text for p in abc.find_all("span",class_="anchor-text")])
print(Con)
#print(abcd)
```
