```python
#to find author ,creator....name
from PyPDF2 import PdfFileReader
 
 
def get_info(path):
    with open(path, 'rb') as f:
        pdf = PdfFileReader(f)
        info = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()
 
    print(info)
 
    author = info.author
    creator = info.creator
    producer = info.producer
    subject = info.subject
    title = info.title
 
if __name__ == '__main__':
    path = "C:\\Users\\saxon\\Desktop\\sss.pdf"
    get_info(path)
```

```python
#converting pdf to text
from tika import parser

raw = parser.from_file("C:\\Users\\saxon\\Desktop\\sss.pdf")
print(raw['content'])
```

```python
#extracting email form pdf
import re
emails = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", raw["content"])
print (emails)
```
